```python
#Testing google download
from google.cloud import storage

def authenticate_implicit_with_adc(project_id="horse-racing-analytics"):
    # Note that the credentials are not specified when constructing the client.
    # Hence, the client library will look for credentials using ADC.
    storage_client = storage.Client(project=project_id)
    print("Authenticated")

def list_blobs():
    """Lists all the blobs in the bucket."""
    bucket_name = "bucket-horseracing-results-pdfs"

    storage_client = storage.Client("horse-racing-analytics")

    # Note: Client.list_blobs requires at least package version 1.17.0.
    blobs = storage_client.list_blobs(bucket_name)

    # Note: The call returns a response only when the iterator is consumed.
    i = 0
    for blob in blobs:
        i += 1
        print(blob.name)
    print(i)
```

```python
authenticate_implicit_with_adc(project_id="horse-racing-analytics")
list_blobs()
```
